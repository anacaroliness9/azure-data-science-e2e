# Read a .csv file from databricks filesystem (dbfs)

```
# File location and type
file_location = "/FileStore/tables/dados_clientes.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
Customer = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)
```

# Take a look in the dataset

```
#display(Customer)
Customer.show(5)
```

## Create a temp table (to be able to query directly using SQL sintaxe)

```
# Create a view or table
temp_table_name = "dados_clientes_csv"
Customer.createOrReplaceTempView(temp_table_name)
```

## Now let`s create a SQL query

```
%sql

/* Query the created temp table in a SQL cell */
select * from `dados_clientes_csv` limit 5
```
